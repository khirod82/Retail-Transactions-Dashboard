```python
import pandas as pd
df = pd.read_csv('/content/Retail_Transaction_Dataset.csv')
df.head()
```

```python
df.shape
```

```python
# Summary statistics for Price, DiscountApplied(%) and TotalAmount
df[['Price', 'DiscountApplied(%)', 'TotalAmount']].describe()
```

```python
# Correlation between Price, DiscountApplied(%) and TotalAmount
df[['Price', 'DiscountApplied(%)', 'TotalAmount']].corr()
```

```python
#Total Amount by Payment Method
df.groupby('PaymentMethod')['TotalAmount'].sum().plot(kind='pie', autopct='%1.1f%%')
```

```python
#Total Amount Spent by Product Category and Payment Method
df.groupby(['ProductCategory', 'PaymentMethod'])['TotalAmount'].sum().unstack().plot(kind='bar')
```

```python
# Distribution of Price
df['Price'].hist(bins=10)
```

```python
# Distribution of TotalAmount
df['TotalAmount'].hist(bins=10)
```

```python
# Scatter plot of Price vs. TotalAmount
df.plot.scatter(x='Price', y='TotalAmount')
```

```python
# Scatter plot of DiscountApplied(%) vs. TotalAmount
df.plot.scatter(x='DiscountApplied(%)', y='TotalAmount')
```
